```python
import json
import os
import copy
import numpy as np
from yolo import YOLO
from PIL import Image
import matplotlib.pyplot as plt
import random
import colorsys
import matplotlib.patches as patches
import cv2
import time
```

# The Binary Classifier

```python
from binary_classification.trainer import Trainer as binary_model
from binary_classification.config import Config
from binary_classification.model import  inception_V3
model=inception_V3(2)
cfg=Config()
binary_model=binary_model(model=model,mode='inference',config=cfg,model_dir='./logs')
binary_model.load_weights("./binary_classification/inception_0046.h5",by_name=True)
```

```python
data_path = "./data/test_0306/"
class_names = ['fefire','blfire','knife','bat','shear']
binary_result_dict = {}
```

```python
for a,b,c in os.walk(data_path):
    for pic in c:
        dir_path = os.path.join(data_path,pic)
        image = Image.open(dir_path)
        image = image.resize((224, 224),Image.ANTIALIAS)
        image = np.array(image,dtype=np.uint8)
        pre=binary_model.detect(image)
        binary_result_dict[pic] = int(np.argmax(pre))
```

# Yolo Detection
### Only detect the retricted picture

```python
yolo_log_dir = 'logs/test_model/'
yolo_model_name = 'test_model_03091144.h5'
yolo_classes_path = 'model_data/danger.txt'
yolo_anchors_path = 'model_data/yolo_anchors.txt'
yolo = YOLO(model_path=os.path.join(yolo_log_dir,yolo_model_name),
            classes_path=yolo_classes_path)

result_dict = {'results':[]}
result = []
```

```python
def get_one_result(pic,SavePic=False):
    image = Image.open(os.path.join(data_path,pic))
    box_result,score_result = yolo.detect(image)
    if SavePic:
        img = cv2.imread(os.path.join(data_path,pic))
    r_dict = {}
    r_list = []
    r_dict['filename']=pic
    for i in range(len(box_result)):
        x1,y1,x2,y2,cls = box_result[i]
        score = score_result[i]
        r_list.append({'xmin':x1,'xmax':x2,'ymin':y1,'ymax':y2,'label':cls+1,'confidence':score})
        if SavePic:
            cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)
#             print("%s %.3f"%(class_names[cls],score_result[i]))
            cv2.putText(img, "%s %.3f"%(class_names[cls],score_result[i]), (x1, y1 - 1), cv2.FONT_HERSHEY_PLAIN, 1.5, (255, 255, 255),2)
            cv2.imwrite(os.path.join("./result",pic),img)
    r_dict['rects']=r_list
    return r_dict

def get_result(binary_result,SavePic=False):
    results = []
    for file in binary_result.keys():
        image = Image.open(os.path.join(data_path,file))
        if not binary_result[file]:
            results.append(get_one_result(file,SavePic=SavePic))
        else:
            print("ignore:{}".format(file))
            if SavePic:
                img = cv2.imread(os.path.join(data_path,file))
                print(file.split('/')[-1])
                cv2.imwrite(os.path.join("./result",file),img)
            results.append({'filename':file,'rects':[]})
    return results
```

```python
class MyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        else:
            return super(MyEncoder, self).default(obj)
# 存储json文件
time_str = time.strftime('%m_%d_%H_%M',time.localtime(time.time()))
json_str = json.dumps(result_dict,cls=MyEncoder)
with open('./result_upload/result_{}.json'.format(time_str), 'w') as json_file:
    json_file.write(json_str)
```
